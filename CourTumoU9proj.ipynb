```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the Transfermarkt page
base_url = 'https://www.imdb.com/title/tt2560140/episodes/?season={}&ref_=ttep'

# Initialize an empty list to store all data
all_data = []

# Set up headers to mimic a real browser request (optional)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

unknown_page_url = None

# Loop through all 4 pages
for page in range(1, 5):
    url = base_url.format(page)

    # Request the page content
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table
    episodes = soup.find_all('div', {'class': 'sc-c68fa612-4 hdCUHo'})

    for div in episodes:
        # Find the rating
        rating_span = div.find('span', class_='ipc-rating-star--rating')
        # Find the vote count
        vote_span = div.find('span', class_='ipc-rating-star--voteCount')

        rating = rating_span.text.strip() if rating_span else ''
        vote_count_raw = vote_span.text.strip() if vote_span else ''

        # Clean vote count
        for vote_count in vote_count_raw:
            # Remove parantheses, turn from 40K format to 40000
            vote_count = int(vote_count_raw[1:-1].replace('K', '')) * 1000

        # Append as a tuple (row)
        all_data.append(('1' ,page, rating, vote_count))

# Convert the list to a DataFrame
df = pd.DataFrame(all_data, columns=['Show', 'Season', 'Rating', 'Total votes'])

# Set display options to show everything in one line
pd.set_option('display.width', None)  # Auto-detect width
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.expand_frame_repr', False)  # Prevent auto-wrapping

# Save the DataFrame to CSV
df.to_csv('attack-on-titan-ratings.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL of the Transfermarkt page
base_url = 'https://www.imdb.com/title/tt0434706/episodes/?season={}&ref_=ttep'

# Initialize an empty list to store all data
all_data = []

# Set up headers to mimic a real browser request (optional)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

unknown_page_url = None

# Loop through all 4 pages
for page in range(0, 2):
    url = base_url.format(page)

    # Request the page content
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table
    episodes = soup.find_all('div', {'class': 'sc-c68fa612-4 hdCUHo'})

    for div in episodes:
        # Find the rating
        rating_span = div.find('span', class_='ipc-rating-star--rating')
        # Find the vote count
        vote_span = div.find('span', class_='ipc-rating-star--voteCount')

        rating = rating_span.text.strip() if rating_span else ''
        vote_count_raw = vote_span.text.strip() if vote_span else ''

        # Clean vote count
        for vote_count in vote_count_raw:
            # Remove parantheses, turn from 40K format to 40000
            try:
                vote_count = int(vote_count_raw[1:-1].replace('K', '')) * 1000
            except ValueError:
                vote_count = int(vote_count_raw[1:-1].replace('K', '').replace('.', '')) * 100

        # Append as a tuple (row)
        all_data.append(('2', page, rating, vote_count))

# Convert the list to a DataFrame
df = pd.DataFrame(all_data, columns=['Show', 'Season', 'Rating', 'Total votes'])

# Save the DataFrame to CSV
df.to_csv('monster-ratings.csv', index=False)
```
